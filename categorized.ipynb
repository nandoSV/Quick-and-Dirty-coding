```python
from urllib import parse
import xmltodict, json
```

```python
with open(f"/content/drive/MyDrive/bestbuy/product_data/categories/{cat_file}", 'r') as cat:
  CatData = xmltodict.parse(cat.read())
  data = json.loads(json.dumps(CatData))
  categories = data["categories"].get("category")
```

```python
for x in range(5):
  levels = ['category_l1', 'category_l2',
            'category_l3', 'category_l4',
            'category_l5']
  train[levels] = ''
  test[levels] = ''
```

```python
test.head()
```

```python
train.head()
```

```python
def bestbuy_category(index:int):
  level = ''
  if index == 0:
    level = "category_l1"
  elif index == 1:
    level = "category_l2"
  elif index == 2:
    level = "category_l3"
  elif index == 3:
    level = "category_l4"
  elif index == 4:
    level = "category_l5"
  return level
```

```python
for cat in categories[:50]:
    path = cat.get("path")
    id = cat.get("id")
    address = path.get("category")

    counter = 0
    index = 0

    for i in range(len(address)):
      lvl = bestbuy_category(counter)
      
      try:
        cat_id =  address[i]["id"]

        if index <= 5: 
          train.loc[train['category'] == id, lvl] = cat_id
          test.loc[test['category'] == id, lvl] = cat_id
          index = index + 1
          counter = counter + 1

        if cat_id == id:
          for i in range(5 - counter):
            lvl = bestbuy_category(counter)
            if index <= 5: 
              train.loc[train['category'] == id, lvl] = ""
              test.loc[test['category'] == id, lvl] = ""
              index = index + 1

      except KeyError as key:
        print(key)
```
